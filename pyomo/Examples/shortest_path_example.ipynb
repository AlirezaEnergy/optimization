```python
import pyomo.environ as pyo
```

## Find the shortest path between node one and 7.

```python
model = pyo.ConcreteModel(name = "Shortest Path")
```

```python
model.x12 = pyo.Var(within=pyo.Binary)
model.x13 = pyo.Var(within=pyo.Binary)

model.x23 = pyo.Var(within=pyo.Binary)
model.x24 = pyo.Var(within=pyo.Binary)
model.x25 = pyo.Var(within=pyo.Binary)

model.x32 = pyo.Var(within=pyo.Binary)
model.x34 = pyo.Var(within=pyo.Binary)
model.x35 = pyo.Var(within=pyo.Binary)
model.x36 = pyo.Var(within=pyo.Binary)

model.x45 = pyo.Var(within=pyo.Binary)
model.x47 = pyo.Var(within=pyo.Binary)

model.x54 = pyo.Var(within=pyo.Binary)
model.x56 = pyo.Var(within=pyo.Binary)
model.x57 = pyo.Var(within=pyo.Binary)

model.x63 = pyo.Var(within=pyo.Binary)
model.x65 = pyo.Var(within=pyo.Binary)
model.x67 = pyo.Var(within=pyo.Binary)
```

```python
model.obj = pyo.Objective(expr = 5*model.x12 + 3*model.x13 + 4*model.x23 + 1*model.x24 + 7*model.x25 + 6*model.x32 +
                          5*model.x34 + 1*model.x35 + 2*model.x36 + 9*model.x45 + 4*model.x47 + 2*model.x54 + 
                          5*model.x56 + 8*model.x57 + 3*model.x63 + 7*model.x65 + 3*model.x67, sense = pyo.maximize)
```

```python
model.BN1 = pyo.Constraint(expr=  1                                             == model.x12 + model.x13)

model.BN2 = pyo.Constraint(expr=  model.x12 + model.x32                         == model.x23 + model.x24 + model.x25)

model.BN3 = pyo.Constraint(expr=  model.x13 + model.x23 + model.x63             == model.x32+model.x34+model.x35+model.x36)

model.BN4 = pyo.Constraint(expr=  model.x24 + model.x34 + model.x54             == model.x45 + model.x47)

model.BN5 = pyo.Constraint(expr=  model.x25 + model.x35 + model.x45 + model.x65 == model.x54 + model.x56 + model.x57)

model.BN6 = pyo.Constraint(expr=  model.x36 + model.x56                         == model.x63 + model.x65 + model.x67)

model.BN7 = pyo.Constraint(expr=  model.x47 + model.x57 + model.x67             == 1)
```

```python
opt = pyo.SolverFactory("glpk")
results = opt.solve(model)
print(pyo.check_optimal_termination(results))
```
